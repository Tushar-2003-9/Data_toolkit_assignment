***Question 8. What does the term “vectorized operation” mean in NumPy?***

Instead of processing elements one by one, NumPy applies the operations to the whole array internally (using optimized C code).

Question 4. How do you calculate the correlation matrix using Seaborn and visualize it with a heatmap?

```python
data = {
    "Sales": [100,200,300,400,500],
    "Profit": [20,40,50,65,80],
    "Marketing_Spend": [10,20,30,40,50]
}

df = pd.DataFrame(data)

corr_matrix = df.corr()
print(corr_matrix)

# Create the graph

plt.figure(figsize = (6,4))
sns.heatmap(corr_matrix, annot = True, cmap = "coolwarm")
plt.title("Correlation Matrix Heatmap")
plt.show()
```

Question 8. Create a line plot with multiple lines using Matplotlib

Question 9.  Generate a Pandas DataFrame and filter rows where a column value is greater than a threshold?

```python
data = {
    "Name" : ["A","B","C","D","E","F"],
    "Marks" : [45,78,85,65,25,65]
}

df = pd.DataFrame(data)

print("Original DataFrame")
print(df)

# Suppose that threshold = 60

threshold = 60

filter_value = df[df["Marks"] > threshold]

print("\nFiltered DataFrame (Marks > 60):")
print(filter_value)
```

Question 11. Perform matrix multiplication using NumPy?

```python
# create two matrix
a = np.array([[1,2,3],[4,5,6]])

b = np.array([[4,5,6],[7,8,9]])

# Multiplication matrix

mul = np.dot(a,b.T)

# Print Matrix
print("Matrix: \n",a)
print("Matrix: \n",b)
print("Result of a* b :\n", mul)
```
